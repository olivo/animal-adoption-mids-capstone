```python
# Detecting blurriness in images.
#!pip install opencv-python
import cv2
from IPython.display import Image, display
import os
import pandas as pd 

def variance_of_laplacian(image):
    return cv2.Laplacian(image, cv2.CV_64F).var()
```

```python
# Sample sharp image.
imagePath = "data/images/0008c5398-1.jpg"
image = cv2.imread(imagePath)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
fm = variance_of_laplacian(gray)
display(Image(filename=imagePath))
print("Laplacian Variance:",fm)
```

```python
# Sample blurry image.
imagePath = "data/images/0008c5398-2.jpg"
image = cv2.imread(imagePath)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
fm = variance_of_laplacian(gray)
display(Image(filename=imagePath))
print("Laplacian Variance:",fm)
```

```python
# Function for computing Laplacian.
def compute_laplacian(imagePath):
    image = cv2.imread(imagePath)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    return variance_of_laplacian(gray)
```

```python
# Iterating through all train images.
pet_ids = []
avg_laplacians = []
first_laplacians = []
min_laplacians = []
max_laplacians = []

pet_id_to_laplacian = dict()

path = 'data/petfinder/train_images'
fileList = os.listdir(path)

for i in range(len(fileList)):
    print("Analyzing image ", i, "out of", len(fileList))
    filename = fileList[i]
    # Applying analysis only on first image for now.
    pet_id, image_number = filename.split(".")[0].split("-")
    
    if pet_id not in pet_id_to_laplacian:
        pet_id_to_laplacian[pet_id] = []
    
    laplacian = compute_laplacian(os.path.join(path + '/' + filename))
    
    # Add the first laplacian.
    if image_number == "1":
        first_laplacians.append(laplacian)
    
    pet_id_to_laplacian[pet_id].append(laplacian)

# Compute average laplacians.
for pet_id in pet_id_to_laplacian:
    pet_ids.append(pet_id)
    avg_laplacians.append(sum(pet_id_to_laplacian[pet_id]) / len(pet_id_to_laplacian[pet_id]))
    min_laplacians.append(min(pet_id_to_laplacian[pet_id]))
    max_laplacians.append(max(pet_id_to_laplacian[pet_id]))
```

```python
# Create dataframe with pet ids and laplacians.
d = {"PetID": pet_ids, "FirstLaPlacianVariance": first_laplacians, "AvgLaPlacianVariance": avg_laplacians, "MinLaPlacianVariance": min_laplacians, "MaxLaPlacianVariance": max_laplacians}
df = pd.DataFrame(d)
```
